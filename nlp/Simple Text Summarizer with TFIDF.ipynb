```python
from nltk.tokenize import sent_tokenize
from nltk.tokenize import TreebankWordTokenizer
import pandas as pd
treeBankWordTokenizer = TreebankWordTokenizer()
sentences = sent_tokenize(tldr_text.lower())
number_of_sentences = len(sentences)

data = {
    'sentence': [],
    'word': [],
    'count': [],
    'total_words': [],
    'term_frequency': []
}
for sentence in sentences:
    words = treeBankWordTokenizer.tokenize(sentence)
    for word in words:
        pass
    
for sentence in sentences:
    words = treeBankWordTokenizer.tokenize(sentence)
    word_unique_counter = {}
    number_of_words = len(words)
    for word in words:
        if word not in word_unique_counter:
            word_unique_counter[word] = 0
        word_unique_counter[word] = word_unique_counter[word] + 1
    
    for word, count in word_unique_counter.items():
        data['sentence'].append(sentence)
        data['word'].append(word)
        data['count'].append(count)
        data['total_words'].append(number_of_words)
        data['term_frequency'].append(count/number_of_words)

df = pd.DataFrame (data, columns = ['sentence','word','count','total_words','term_frequency'])
df
```
